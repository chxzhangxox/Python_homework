<h3>The Complex Exponential</h3>

You don't need to understand the precise meaning of the Fast Fourier Transform in order to use it, but a little understanding can be helpful. For instance, why does the FFT produce an array of complex numbers, and how can the real and complex parts be interpreted? When the "power" is plotted against frequency, why is the plot symmetric about the line of zero frequency? 

To explore questions like these, we need to understand the complex exponential function,

$$z(t) = e^{2 \pi j \omega t},$$

where $j = \sqrt{-1}$ (yes, we're using the letter $j$ to be the imaginary unit instead of another common convention, $i$), $t$ can be interpreted as time (in many applications), and $\omega$ can then be interpreted as frequency in cycles per unit of time (instead of rad/s due to the factor of $2 \pi$).  

Let's start by plotting this function. The code looks long, but most of it is for fancy formatting.

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from mpl_toolkits.mplot3d.art3d import Line3D
%matplotlib inline

w = 1 # instead of omega
t = np.linspace(0,2,1000)
z = np.exp(2j*np.pi*w*t)

# These set the font and enable LaTeX to produce pretty math
plt.rc('text', usetex=True)
plt.rc('font', family='serif')
plt.rc('font', size=16) 

fig = plt.figure()
fig.set_size_inches((8,6))
ax = fig.add_subplot(111, projection='3d')
ax.plot(t, np.real(z), zs= np.imag(z))

# If you can't figure out what these do, try commenting them out to see the difference
ax.set_xlabel("$t$")
ax.set_ylabel("$\Re(z)$")
ax.set_zlabel("$\Im(z)$")
ax.set_title("$z(t) = e^{2 \pi j \omega t}, \omega = 1$")
ax.xaxis.labelpad =10            # spaces the label text from the tick labels
ax.yaxis.set_rotate_label(False) # otherwise the label text is rotated 
ax.yaxis.labelpad =15
ax.zaxis.set_rotate_label(False)
ax.zaxis.labelpad =7

ax.add_line(Line3D([0,0],[0,0],[0,1],color='k'))
ax.add_line(Line3D([0,0],[0,1],[0,0],color='k'))
ax.add_line(Line3D([0,np.max(t)],[0,0],[0,0],color='k'))

ax.set_xticks([0,0.5,1,1.5,2])
ax.set_yticks(np.arange(-1,1.1,0.5))
ax.set_zticks(np.arange(-1,1.1,0.5))
```

We need a 3D plot because $z(t)$ has both real and imaginary parts. Does it look familiar? In case it doesn't, let's look at the real part $\Re(z)$ and imaginary part $\Im(z)$ separately.

```python
plt.plot(t,np.real(z))
plt.xlabel("$t$")
plt.ylabel("$\Re(z)$")
plt.title("$\Re(z(t))$")
```

Looks an awful lot like the $\cos$ function to me.

```python
plt.plot(t,np.imag(z))
plt.xlabel("$t$")
plt.ylabel("$\Im(z)$")
plt.title("$\Im(z(t))$")
```

And that's a $\sin$ wave if I ever saw one.

This all follows from that beautiful equation, <a href = "https://en.wikipedia.org/wiki/Euler%27s_formula">Euler's formula</a>, that you developed in early calculus while working with power series:

$$ e^{jt} = \cos(t) + j \sin(t)$$

In other words, the complex exponential can be interpreted as the sum of a real cosine wave and an imaginary sine wave.

<h3>The FFT (by example)</h3>

Now for the FFT. Let's say we have a signal:

$$x(t) = \cos(2 \pi t) + \frac{1}{4} \sin(4 \pi t)$$

It's not so hard to see that this is the sum of two _sinusoids_ (the general term for sine or cosine functions)  of different amplitudes and frequencies, i.e.

$$x(t) = x_1(t) + x_2(t)$$

where

$$x_1(t) = a_1 cos(2 \pi \omega_1 t)$$ 

with $a_1 = 1$ and $\omega_1 = 1$; and 

$$x_2(t) = a_2 sin(2 \pi \omega_2 t)$$ 

with $a_2 = \frac{1}{4}$ and $\omega_2 = 2$.

The component functions appear separately in the plot below.

```python
t_fine = np.linspace(0,1,10000)
a1, w1, a2, w2 = 1., 1., 0.25, 2.
x1 = a1*np.cos(2*np.pi*w1*t_fine)
x2 = a2*np.sin(2*np.pi*w2*t_fine)
plt.plot(t_fine,x1,t_fine,x2)
plt.legend(("$x_1(t)$", "$x_2(t)$"))
plt.xlabel("$t$")
plt.ylabel("$x_i(t)$")
```

Together, their sum looks like this.

```python
x_fine = x1 + x2
plt.plot(t_fine,x_fine)
plt.xlabel("$t$")
plt.ylabel("$x(t)$")
```

Because the FFT is just a _fast_ algorithm for the _discrete_ Fourier transform, the FFT works with _samples_ of such a function, that is, the values of the function at equally spaced points in time. We generated the plots above using samples, actually, but the plots look quite smooth because we took 10000 samples and connected them with lines. If $t$ represents time in seconds, we would say that the sampling frequency $f$ is 10000 samples per second, or the sampling period $\delta$ is 1/10000 of a second.

However, to emphasize the discrete nature and to make the information to follow easier to interpret, let's take onle $N = 10$ samples over the course of one second, that is, we'll use a much lower sampling frequency $f = 10$ samples per second and (equivalently) a longer sampling period $\delta = 0.1$ second.

```python
N = 10
t_max = 1
f = N / t_max           # sampling frequency
dt = 0.1                # sampling period delta
t = np.arange(0,1,dt)
n = np.arange(N)
x1 = np.cos(t*2*np.pi)
x2 = 0.25*np.sin(2.0*t*2*np.pi)
x = x1 + x2

fig = plt.figure()
ax1 = fig.add_subplot(111)
ax1.set_xticks(np.arange(N+1))
ax1.set_xlabel("$n$")

ax2 = ax1.twiny()
ax2.plot(t_fine, x_fine, "-", t, x, "o")
ax2.set_xlim([0,1])
ax2.set_xlabel("$t$")

ax2.legend(["$x(t)$", "$x[n]$"])
```

Here we have 10 samples of the underlying function equally spaced with sampling period $\delta = 0.1$s. The different notation $x[n]$ emphasizes that these are only samples of the function $x(t)$ at discrete points; $n$ takes on only integer values $0, 1, \dots, N-1$ and $x[n] = x(t = n \delta) = x\left(t = \frac{n}{f} \right)$.

According to the Scipy documentation, the FFT takes these samples $x[n]$ and finds coefficients $y[k]$ for $k = 0, 1, \dots, N-1$ such that

$$ x[n ]= \frac{1}{N} \sum_{k=0}^{N-1} y[k] e^{2\pi j \frac{kn}{N}} $$



Let's take that apart:
* $x[n]$ are the values of the function $x(t)$ at the equally spaced sample points.
* $y[k]$ are the coefficients the FFT is going to find for us. Not sure what they're good for yet. Not really sure what $k$ is yet, other than an index.
* $e^{2\pi j \frac{kn}{N}}$ look like the complex exponentials $e^{2 \pi j \omega t}$, but it's a function of index $n$ instead of time $t$, and the frequency is $\frac{k}{N}$ instead of $\omega$. Weird.

Maybe expanding that a bit will help:

$$ x[n] = \frac{y[0]}{N} e^{2 \pi \frac{0}{N}n} + \frac{y[1]}{N} e^{2 \pi \frac{1}{N}n} + \dots + \frac{y[N-1]}{N} e^{2 \pi \frac{(N-1)}{N}n}  $$

The FFT allows us to write the sample values $x[n]$ as the sum of the values of complex exponentials in $n$. If we have a lot of sample points, this is like writing the function itself as the sum of complex exponential functions, which can be thought of as sines and cosines. The FFT is breaking the function apart into sinusoidal components of different frequencies.

Associated with each coefficient $y[k]$ is a term $e^{2\pi j \frac{k}{N}n}$. Since the integers $n$ correspond with each of the sample times $t = n \delta = \frac{n}{f}$, we can think of $e^{2\pi j \frac{k}{N}n}$ as the value of a complex exponential function $e^{2 \pi j \left(\frac{k}{N}\right) ft}$ at a particular time $t = \frac{n}{f}$. In other words, this is a complex exponential $e^{2 \pi j \omega t}$ with $\omega = \frac{k}{N} f$, a fraction of the sampling frequency. The FFT is breaking down the original signal into components of frequencies $0, \frac{1}{N}f, \frac{2}{N}f, \dots, \frac{N-1}{N} f$. So in a sense, we're approximating 

$$ x(t) \approx \frac{y[0]}{N} e^{2 \pi \frac{0}{N}ft} + \frac{y[1]}{N} e^{2 \pi \frac{1}{N}ft} + \dots + \frac{y[N-1]}{N} e^{2 \pi \frac{(N-1)}{N}ft} = \frac{1}{N} \sum_{k=0}^{N-1} y[k] e^{2\pi j \frac{k}{N}ft}$$

in the sense that the relationship will be exactly true at the sample points. 

Let's take a look at some of these complex exponentials in the case of $N = 10$ samples at a frequency $f = 10$ samples per second.



```python
def plot_exp(k, ri):
    f = 1/dt
    z = np.exp(2j*np.pi*k*f*t/N)
    z_fine = np.exp(2j*np.pi*k*f*t_fine/N)
    plt.plot(t,ri(z),'o',t_fine, ri(z_fine), "-")
    plt.title("$e^{2 \pi j " + str(k) + "/N f t}$")
    # plt.legend(["$[n]$", "$x(t)$"])
    
k = 3        # try values k = 0, 1, ..., N-1
ri = np.real # you can try np.imag to see the imaginary part
plot_exp(k, np.real)
```

The term for $k = 0$ is just a constant.

```python
plot_exp(0,np.real)
```

The term $k = 1$ is a slow wave

```python
plot_exp(1,np.real)
plot_exp(1,np.imag)
plt.legend(["","real part","","imaginary part"])
```

Look at $k = 5$, that is, $k = \frac{N}{2}$.

```python
plot_exp(5,np.real)
```

Note that the samples can't possibly oscillate faster than this. If we were to try, look what happens:

```python
plot_exp(6,np.real)
```

At the discrete sample points, this looks identical to the slightly slower $k=4$.

```python
plot_exp(6,np.real)
plot_exp(4,np.real)
plt.legend(["","$k = 6$","","$k = 4$"])
```

$k = 7$ looks the same as an even slower wave (at the sample points).

```python
plot_exp(7,np.real)
plot_exp(3,np.real)
plt.legend(["","$k = 7$","","$k = 3$"])
```

Actually, it looks like the samples of the real parts for $k = N/2 + i$ are the same as the samples of the real parts for $k = N/2 -i$. 

```python
i = 4  # try any integer 0-N/2
plot_exp(N/2-i, np.real)
plot_exp(N/2+i, np.real)
```

How about the imaginary parts?

```python
i = 3  # try any integer 0-N/2
plot_exp(N/2-i, np.imag)
plot_exp(N/2+i, np.imag)
```

Look carefully - the imaginary components have the same magnitude, but the sign is flipped.

It turns out that because $\cos$ is an even function, that is, $\cos(-a) = \cos(a)$, and because $\sin$ is an odd function, $\sin(-a) = -\sin(a)$, we can get both the real and imaginary parts to match using negative frequencies. That is, the vaues of samples at  $k = N/2 + i$ are the same as the values of the samples for $k = -N/2 + i$. 

```python
i = 3 # try any integer 0-N/2
plot_exp(-N/2+i, np.imag)
plot_exp(N/2+i, np.imag)
```

```python
i = 3  # try any integer 0-N/2
plot_exp(-N/2+i, np.real)
plot_exp(N/2+i, np.real)
```

The frequency corresponding with $k = N/2$ is $\frac{N/2}{N}f = \frac{f}{2}$. It's called the Nyquist frequency, and there are a few things to be said about it:
* Sampling at frequency $f$, sines and cosines of frequency $\frac{f}{2} + \frac{i}{N}f$  look the same as sines and cosines of frequency $-\frac{f}{2} + \frac{i}{N}f$
* If you are trying to break down a signal into component frequencies, you can only really break it down into frequencies lying between $-\frac{f}{2}$ and $\frac{f}{2}$; frequencies higher than $\frac{f}{2}$ each look the same as a corresponding negative frequency
* If you want to represent a signal (like an audio signal) by taking samples, your Nyquist frequency had better be faster than the (highest important) frequency of the signal, or your samples can't do justice to the complexity of the original signal. 

All this was to explain that instead of looking at the FFT as breaking the signal into frequencies $0, \frac{1}{N}f, \frac{2}{N}f, \dots, \frac{N-1}{N} f$,  we can also think of it as breaking the signal into frequencies $-\frac{f}{2}, \frac{N/2+1}{N} f, \dots, 0, \frac{1}{N}f, \dots, \frac{N/2-1}{N} f$.

Finally, we can begin to appreciate what the FFT is telling us. Let's take the FFT of our signal, which I'll plot again for convenience.

```python
dt = 0.1                # sampling period delta
f = 1/dt                # sampling frequency
t = np.arange(0,1,dt)
N = len(t)              # number of samples
n = np.arange(N)
x1 = np.cos(t*2*np.pi)
x2 = 0.25*np.sin(2.0*t*2*np.pi)
x = x1 + x2
plt.plot(t,x,'o',t_fine, x_fine, "-")
plt.legend(["$x[n]$", "$x(t)$"])
plt.title("Original Signal")
```

```python
from scipy.fftpack import fft, ifft, fftfreq
from IPython.display import display
y = fft(x)
display(y)
```

So the FFT takes in $N$ sample points of our original signal and gives an array of $N$ complex numbers. Each of these complex numbers is the coefficient $y[k]$ that goes on a complex exponential of frequency $\frac{k}{N}f$. When we sum all these complex exponentials up, we should get back our original signal.

Let's take a look at the coefficients next to all the frequencies they correspond with.

```python
print("{: >40} {: >15}".format("y[k]", "Frequency"))
for k in range(N):
    print("{: >40} {: >15}".format(str(y[k]), str(float(k)/N*f)))
```

Some of these numbers - the ones with exponents $10^{-15}$ and $10^{-16}$ are really tiny. I can tell you they're just caused by numerical precision issues. Let's get rid of them.

```python
for k in range(N):
    if np.abs(y[k]) > 1e-13:
        a = np.round(np.real(y[k]),3)
        b = np.round(np.imag(y[k]),3)
        print str(a + b*1j),"corresponds with frequency", str(float(k)/N*f), "samples per second"
```

Interesting. This is telling us that our signal - which we know has a real $\cos$ at frequency $1.0$ and $0.25$ of a real $\sin$ at frequency $2$ is really made up of four parts, some of which have imaginary coefficients. What's going on? How can this be right? Let's sum up those complex exponentials and see if we get back our original signal.

```python
x2_fine = np.zeros(x_fine.shape)
for k, yk in enumerate(y):
    x2_fine = x2_fine  + 1.0/N * yk * np.exp(2j * np.pi * float(k)/N * f * t_fine)
plt.plot(t_fine, np.real(x2_fine))
plt.title("Sum of Complex Exponentials")
```

Aside: I did that calculation in a loop, but I don't need to. I can use broadcasting:

```python
k = np.arange(N)
x2_fine = np.sum(1.0/N * y * np.exp(2j * np.pi * k/N * f * t_fine[:,np.newaxis]),1)
plt.plot(t_fine, np.real(x2_fine))
plt.title("Sum of Complex Exponentials")
```

That doesn't look like our original signal. 

```python
plt.plot(t_fine, np.real(x2_fine))
plt.plot(t_fine, np.real(x_fine))
plt.legend(["sum of complex exponentials","original signal"])
plt.title("Sum of Complex Exponentials vs Original Signal")
```

Look again:

```python
plt.plot(t_fine, np.real(x2_fine))
plt.plot(t_fine, np.real(x_fine))
plt.plot(t, np.real(x),'o')
plt.title("Sum of Complex Exponentials vs Original Signal\n (at discrete samples points)")
```

So actually, it is exactly our original signal - but only at the sample points!

But how can that be? What about the imaginary parts?

```python
plt.plot(t_fine, np.imag(x2_fine))
plt.plot(t_fine, np.imag(x_fine))
plt.plot(t, np.imag(x),'o')
plt.title("Sum of Complex Exponentials vs Original Signal\n (imaginary parts)")
```

At the sample points, all the imaginary parts went to zero. Whoa. Well... why? How can we make sense of it?

Well, first, remember that those higher frequencies can really be converted to negative lower frequencies. The `fftfreq` function does that for us. We just have to provide $N$ and $\delta = \frac{1}{f}$. It calculates the frequencies $\frac{N/2+1}{N} f, \dots, 0, \frac{1}{N}f, \dots, \frac{N/2-1}{N} f$ and puts them in the right order to correspond with our cooefficients $y[k]$, that is $0, \frac{1}{N}f, \dots, \frac{N/2-1}{N}, -\frac{f}{2}, \frac{-N/2+1}{N} f, \dots,  \frac{-1}{N}f$

```python
w = fftfreq(N,dt)
display(w)
```

```python
for k in range(N):
    if np.abs(y[k]) > 1e-13:
        a = np.round(np.real(y[k]),3)
        b = np.round(np.imag(y[k]),3)
        print str(a + b*1j),"corresponds with frequency", str(w[k]), "samples per second"
```

Hmm, now we're getting somewhere.  Using the negative frequencies, we get:

```python
x2_fine = np.sum(1.0/N * y * np.exp(2j * np.pi * w * t_fine[:,np.newaxis]),1)
plt.plot(t_fine, np.real(x2_fine),"-")
plt.plot(t_fine, np.real(x_fine),"--")
plt.title("Sum of Complex Exponentials (using negative frequencies) \n vs Original Signal")
```

```python
plt.plot(t_fine, np.imag(x2_fine),"-")
plt.plot(t_fine, np.imag(x_fine),"--")
plt.title("Basically zero everwhere")
```

Which, in this case, is exactly our original signal at _all_ points. You're not going to get perfect results like this, as not all functions can be represented in terms of frequencies $\frac{f}{2}, \frac{N/2+1}{N} f, \dots, 0, \frac{1}{N}f, \dots, \frac{N/2-1}{N} f$. But you will always get the right values at the sample points.

Let's get just a little more intuition, if we can, by trying to figure out _how_ the imaginary parts cancelled out. Let's look at the real and imaginary parts of the FFT cooeficients separately.

```python
plt.plot(w,np.real(y),'o')
plt.xlabel("Frequency (samples per second)")
plt.ylabel("$\Re(y[k])$")
plt.title("FFT Coefficients (real part) at each frequency")
```

This is saying that the FFT cooeficients are 5 on the complex exponentials with frequencies $+1$ and $-1$. Let's look at the real and imaginary parts of the complex exponentials with those frequencies:

```python
def plot_exp(w, ri, spec):
    z = np.exp(2j*np.pi*w*t)
    z_fine = np.exp(2j*np.pi*w*t_fine)
    plt.plot(t,ri(z),'o',t_fine, ri(z_fine), spec)
     # plt.title("$e^{2 \pi j " + str(w) + " t}$")
plot_exp(1,np.real, "-")
plot_exp(-1,np.real, "--")
plt.title("Complex Exponentials with Frequencies $+1$, $-1$ (real part)")
```

The real parts are the same, as we would expect,  as $\cos$ is an even function and thus $\cos(-a) = \cos(a)$. The imaginary parts?

```python
plot_exp(1,np.imag, "-")
plot_exp(-1,np.imag, "--")
plt.title("Complex Exponentials with Frequencies $+1$, $-1$ (imaginary part)")
```

Of course, as $\sin$ is odd, so $\sin(-a) = \sin(a)$.

So with a real coefficient of 5 on the complex exponential with frequencies $1$ and $-1$, the imaginary parts will cancel out and the real parts will add.

$$ 5 e^{2 \pi j \left(1\right) t} + 5 e^{2 \pi j \left(-1\right) t} = 5 \cos(2 \pi t) + 5j\sin(2 \pi t) + 5 \cos(2 \pi t) - 5j\sin(2 \pi t) = 10 \cos(2 \pi t)$$

Why is there $10\cos(t)$ when the original signal just has $1.0\cos(t)$?  Look back to the beginning:
$$ x[n ]= \frac{1}{N} \sum_{k=0}^{N-1} y[k] e^{2\pi j \frac{kn}{N}} $$
Everything gets multiplied by $\frac{1}{N}$, which cancels out that factor of 10.

_So we see that if the original signal is real and has a component $a \cos(\omega)$, then the Fourier coefficients corresponding with that frequency will have real part $a\frac{N}{2}$._

Now let's take a look at the imaginary part of the Fourier coefficients.

```python
plt.plot(w,np.imag(y),'o')
plt.xlabel("Frequency (samples per second)")
plt.ylabel("$\Re(y[k])$")
plt.title("FFT Coefficients (imaginary part) at each frequency")
```

This is saying that the FFT cooeficients is $1.25j$ on the complex exponential with frequency $-2$ and the FFT coefficient is $-1.25j$ on the complex exponential with frequency $+2$. Let's look at the real and imaginary parts of the complex exponentials with those frequencies:

```python
plot_exp(2,np.real, "-")
plot_exp(-2,np.real, "--")
plt.title("Complex Exponentials with Frequencies $+2$, $-2$ (real part)")
```

As we knew from before, the real (cosine) parts are the same. But the coefficients have opposite signs $(\pm1.25j)$, so the cosine parts will cancel out. The imaginary parts are:

```python
plot_exp(2,np.imag, "-")
plot_exp(-2,np.imag, "--")
plt.title("Complex Exponentials with Frequencies $+2$, $-2$ (imaginary part)")
```

Of course, they're opposite in sign. With coefficients that are opposite in sign, the parts will sum together! Perhaps even more interesting, because the coefficient is imaginary, we'll end up with $j^2 = -1$ that will end up making everything real and positive!

$$ -1.25j e^{2 \pi j \left(2\right) t} + 1.25j e^{2 \pi j \left(-2\right) t} = -1.25j \cos(4 \pi t) + -1.25j^2\sin(4 \pi t) + 1.25j \cos(4 \pi t) - 1.25j^2\sin(4 \pi t) = 2.5\sin(4 \pi t)$$

As with the real parts, everything will get scaled by $\frac{1}{N}$, making the coefficient $0.25$ on the $\sin$ term.

_So we see that if the original signal is real and has a component $b \sin(\omega)$, then the Fourier coefficients corresponding with that frequency will have imaginary part $b\frac{N}{2}$._

And so we've come full circle, as

$$\frac{1}{N} 10 \cos(2 \pi (1) t) + \frac{1}{N} 2.5 \sin(2 \pi (2) t) =  \cos(2 \pi t) + \frac{1}{4} \sin(4 \pi t)$$

like we started with.

<h3>FFT Computation</h3>

I won't go into the FFT algorithm. But for those of you who have taken linear algebra, I'll mention a really important point in the discrete Fourier transform: the family of discrete exponentials $e^{2\pi j \frac{kn}{N}}$ form an orthogonal basis, and the FFT changes our representation of the signal to that basis.

That is, for a given frequency number $k$, you can calculate the value of the complex exponential $e^{2\pi j \frac{kn}{N}}$ at each of the sample points $n = 0, 1, \dots, N-1$. 

```python
k = 1
n = np.arange(N)
def calc_exp(k):
    return np.exp(2j*np.pi*k*n/N)
z = calc_exp(k)
display(z)
```

You can think of this as a vector in $N$ dimensional space of complex vectors. There one of these vectors corresponding with each frequency numbers $k = 0, 1, \dots, N-1$. There are $N$ total vectors so as long as they are linearly independent, they will form a basis that spans the $N$ dimensional space. 

If we have a signal $x[n]$, performing the discrete Fourier transform is simply a change to that basis. The $x[n]$ are the components of the signal in the natural basis, and the $y[k]$ are the components of the same signal in this new basis.

Turns out that not only are the vectors in this new basis linearly independent, they're orthogonal. You can see that their inner product is zero:

```python
# Try any two different integers 0, 1, ..., 9
ka = 1
kb = 2
za = calc_exp(ka)
zb = calc_exp(kb)
display(np.dot(za,np.conj(zb))) # in case you haven't  complex conjugate on  
```

This makes it really easy to change basis - we just take the inner product of the original vector with each of these basis vectors. 

```python
for k in range(N):
    z = calc_exp(k)
    print np.allclose(np.dot(x,np.conj(z)), y[k]) # don't forget the complex conjugate 
```

(Although, it turns out that for large signals - lots of samples - even this is too slow. If we have a signal with $N$ samples, we need to take the inner product with $N$ Fourier basis vectors, and this is $N^2$ operations. The FFT is an algorithm that calculates the coefficients in only $N \log(N)$ operations.)

These basis vectors are not normalized though; the inner product of any with itself is $N$.

```python
for k in range(N):
    z = calc_exp(k)
    print(np.dot(z,np.conj(z))) # don't forget the complex conjugate 
```

That's why we need the factor of $\frac{1}{N}$ when converting back from Fourier basis to the natural basis.
