```python
import matplotlib.pyplot as plt
from scipy.io import wavfile as spwav
import scipy.fftpack as spft 
import numpy as np
%matplotlib inline
```

```python
rate,sig_stereo = spwav.read("wooguy.wav")
print sig.shape # it's stereo - there are two audio channels, each with 695k samples.
print sig.dtype # data type is int16 - we'll make sure to convert the modified data back to this type
```

```python
# Get rid of second channel while we're figuring out which frequencies to zero
sig = sig_stereo[:,0]
```

```python
time_step = 1./rate # time between samples
sample_freq = spft.fftfreq(sig.size, d=time_step) # get the frequencies(just like the tutorial)
print sample_freq.shape
```

```python
sig_fft = spft.fft(sig) # take the fft
print sig_fft.shape
print sig_fft.dtype
```

```python
# Take the absolute value to get the "power" of each frequency
power = np.abs(sig_fft)
```

```python
plt.plot(sample_freq,power)
```

```python
# plot is symmetric, we only need the right half
mask = sample_freq>0
freq = sample_freq[mask]
power = power[mask]
plt.plot(freq,power)
```

```python
# let's szoom in on the important stuff. There's not much going on above 5000 Hz.
mask = freq < 5000
plt.plot(freq[mask],power[mask])
most_powerful_freq = freq[np.argmax(power)]
print most_powerful_freq
```

```python
# Listen to 1700Hz from the tone generator - http://www.szynalski.com/tone-generator/
# 1700 Hz is too high of a pitch for that guy's voice
# Let's look at the second most powerful frequency
mask = freq < 1000
plt.plot(freq[mask],power[mask])
```

```python
# There's a peak from 500 to 600 Hz. 
# The guy's voice sounds like ~575Hz from the tone generator, which is in that range.
# Let's get rid of all those frequence the stereo signal.
# Be sure you to use the original, symmetric sample_freq and sig_fft arrays
mask = np.logical_and(500 < sample_freq, sample_freq < 600)
sig_fft[mask] = 0
plt.plot(sample_freq, np.abs(sig_fft))
```

```python
# Oops, we forgot to keep it symmetric
mask = np.logical_and(-600 < sample_freq, sample_freq < -500)
sig_fft[mask] = 0
plt.plot(sample_freq, np.abs(sig_fft))
```

```python
sig = spft.ifft(sig_fft)
```

```python
# The remaining complex parts are just due to numerical error
print np.max(np.abs(sig.imag))
```

```python
# Get rid of imaginary parts, keep just the real part
sig = sig.real
```

```python
# Convert back to int16 and save to file
sig = np.array(sig, dtype = "int16")
spwav.write("no_wooguy.wav", rate, sig)
```

```python
# That removed Woo Guy from one channel. Let's remove him from the stereo audio.
sig_fft = spft.fft(sig_stereo, axis = 0)
mask = np.logical_and(500 < sample_freq, sample_freq < 600)
sig_fft[mask,:] = 0
mask = np.logical_and(-600 < sample_freq, sample_freq < -500)
sig_fft[mask,:] = 0
sig = spft.ifft(sig_fft, axis = 0)
sig = sig.real
sig = np.array(sig, dtype = "int16")
spwav.write("no_wooguy2.wav", rate, sig)
```
